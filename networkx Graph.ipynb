```python
import networkx as nx
graph = {
    "0": ["1","3","8"],
    "1": ["0","7"],
    "2": ["3","5","7"],
    "3": ["0","2","4"],
    "4": ["3","8"],
    "5": ["2","6"],
    "6": ["5"],
    "7": ["1","2"],
    "8": ["0","4"]
}
G = nx.Graph(graph)
nx.draw(G, with_labels=True)
```

```python
list(nx.bfs_tree(G, '0'))
```

```python
list(nx.dfs_tree(G, '0'))
```

```python
list(nx.dijkstra_predecessor_and_distance(G,"0"))
```

```python
graph_dijkstra = nx.Graph()

graph_dijkstra.add_edge('0','1',weight=8)
graph_dijkstra.add_edge('0','2',weight=9)
graph_dijkstra.add_edge('1','3',weight=1)
graph_dijkstra.add_edge('1','5',weight=7)
graph_dijkstra.add_edge('1','6',weight=9)
graph_dijkstra.add_edge('2','5',weight=6)
graph_dijkstra.add_edge('2','6',weight=4)
graph_dijkstra.add_edge('3','7',weight=4)
graph_dijkstra.add_edge('4','6',weight=7)
graph_dijkstra.add_edge('4','7',weight=7)
graph_dijkstra.add_edge('5','7',weight=2)
graph_dijkstra.add_edge('6','7',weight=6)
```

```python
nx.draw(graph_dijkstra,with_labels=True)
nx.dijkstra_path(graph_dijkstra,"0","4")
```
